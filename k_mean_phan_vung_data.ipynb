1. Phân cụm dữ liệu

```python
#Khai báo thư viện
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
#Chuẩn bị dữ liệu
n_samples = 1000
random_state = 170
center_point = [[1, 1], [-1, -1], [1, -1]] #Sinh ngẫu nhiên các điểm xung quanh vị trí

X, y = make_blobs(n_samples=n_samples, random_state=random_state, centers=center_point, cluster_std= 0.6)
print ("Số chiều dưx liệu: ", X.shape, y.shape)
print("5 điểm dữ liệu đầu tiên: \n", X[:5])
```

```python
#Vẽ các điểm dữu liệu
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c = 'blue')
plt.title('Các điểm dữu liệu trước khi phân cụm. Số lượng: {}'.format(n_samples))
plt.show()
```

```python
#Xây dựng giải thuật K_means và huấn luyện
k_cluster = 3
k_mean_model = KMeans(n_clusters=k_cluster, random_state=random_state)
k_mean_model.fit(X)

centers = np.array(k_mean_model.cluster_centers_)

print("Tâm cụm sau khi training ({} tâm): \n ".format(k_cluster), centers)
```

```python
#Kiểm tra 
y_pred = k_mean_model.predict(X)
print("Kết quả dự đoán cho 5 mẫu dữ liệu đầu tiên trong tập data: \n")
print(y_pred[:5])
plt.figure(figsize=(12, 12))
plt.subplot(222)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.scatter(centers[:, 0], centers[:, 1], c='red')
plt.title("Các điểm dữ liệu sau khi phân cụm.")

plt.show()
```

```python
n_samples = 1000 # Số lượng mẫu
random_state = 170 
center_points = [[1, 1], [-1, 1]] # sinh ngẫu nhiên các điểm xung quanh 2 vị trí tâm cố định

X, y = make_blobs(n_samples=n_samples, random_state=random_state,centers=center_points, cluster_std=0.6)
print("Số chiều dữ liệu: ", X.shape, y.shape)
print("5 điểm dữ liệu đầu tiên: \n", X[:5])
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c= 'blue')
plt.title("Các điểm dữ liệu trước khi phân cụm. Số lượng: {}".format(n_samples))  
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(131)

k_cluster = 3
k_mean_model = KMeans(n_clusters=k_cluster, random_state=random_state)
k_mean_model.fit(X)
centers = np.array(k_mean_model.cluster_centers_)
y_pred = k_mean_model.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.scatter(centers[:, 0], centers[:, 1], c='red')
plt.title("Các điểm dữ liệu sau khi phân 3 cụm.")

plt.subplot(132)

k_cluster = 4
k_mean_model = KMeans(n_clusters=k_cluster, random_state=random_state)
k_mean_model.fit(X)
centers = np.array(k_mean_model.cluster_centers_)
y_pred = k_mean_model.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.scatter(centers[:, 0], centers[:, 1], c='red')
plt.title("Các điểm dữ liệu sau khi phân 4 cụm.")

plt.subplot(133)

k_cluster = 5
k_mean_model = KMeans(n_clusters=k_cluster, random_state=random_state)
k_mean_model.fit(X)
centers = np.array(k_mean_model.cluster_centers_)
y_pred = k_mean_model.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.scatter(centers[:, 0], centers[:, 1], c='red')
plt.title("Các điểm dữ liệu sau khi phân 5 cụm.")

plt.show()
```
