```python
#Import the necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
#Load the dataset
iris = load_iris()
X = iris.data[:,[2,3]]
y = iris.target
```

```python
#split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 1)
```

```python
#Create s decision tree classifier
clf = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state= 1)
```

```python
#Train the classifier on the training set
clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10, 6))
plot_tree(clf, filled=True, rounded=True, class_names=iris.target_names, feature_names=iris.feature_names[2:4])
plt.show()
```

```python
# Dùng cây quyết định này để đưa ra dự đoán trên tập test
y_ = clf.predict(X_test)

# Kiểm tra độ chính xác
# số lượng đoán đúng
n_acc = np.sum(y_ ==y_test)
print('số lượng trường hợp đoán đúng:',n_acc, '/', len(y_test))
# độ chính xác 
acc = n_acc/len(y_test)
print('độ chính xác: ', acc)
```
